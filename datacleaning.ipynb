```python
null_counts = df.isnull().sum()
null_counts_dict = null_counts.to_dict()

null_counts_dict
```

```python
sorted_null_counts = dict(sorted(null_counts_dict.items(), key=lambda item: item[1], reverse=True))

sorted_column_names = list(sorted_null_counts.keys())

sorted_null_counts

# looks like we can delete the following columns for sure:
# pdeath
# psych4
# urine
# glucose
# bloodchem4
```

```python
sex_column = df['sex']

unique_sex_values = set()

for value in sex_column:
    unique_sex_values.add(value)

print(unique_sex_values)
```

```python
df = df[df['sex'] != '1']

df['sex'].replace({'female': '1', 'M': '0', 'Male': '0', 'male': '0'}, inplace=True)

df['sex'] = df['sex'].astype(int)
df
```
